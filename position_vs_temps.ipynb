```python
import numpy as np
import pylab as pl
```

```python
# donn√©es
x = [0, 1.2, 3.5, 9.2, 15.4, 24.8, 36.7]
t = [0, 1, 2 ,3, 4, 5, 6]

# incertitude
yerr = 2

# r√©gression polynomiale
p = np.polyfit(t, x, 2)

# affichage des r√©sultats
print('position initiale', p[2], "m") 
print('vitesse initiale', p[1], "m/s") 
print('acc√©leration', p[0]*2, "m/s^2") 

# graphique
pl.errorbar(t, x, yerr=yerr, ls='', marker='+')
pl.plot(t, np.polyval(p, t))
pl.xlabel('t (s)')
pl.ylabel('x (m)')
pl.title('Position en fonction du temps')
pl.grid()
```
